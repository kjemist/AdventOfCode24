```python
def check_row_diff(row):
    diffs = [j-i for i,j in zip(row[:-1], row[1:])]
    return diffs
```

```python
df = read_df("./reports.csv")
```

```python
safe_lists = []
for row in df.iterrows():
    row_list = [x for x in row[1].to_list() if not math.isnan(x)]
    diffs = check_row_diff(row_list)
    print(row_list, "\n")
    if (all(0 < abs(i) < 4 for i in diffs)) and (all(i > 0 for i in diffs) or all(i < 0 for i in diffs)):
        print("Safe list!")
        print(row_list)
        print(diffs, "\n")
        safe_lists.append(row)
```
