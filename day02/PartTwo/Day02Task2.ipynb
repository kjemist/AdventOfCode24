```python
import pandas as pd
import math
```

```python
def read_df(path):
    df = pd.read_csv(path, header=None, sep = " ", names=[1,2,3,4,5,6,7,8])
    print(df)
    return df
```

```python
def check_row_diff(row):
    diffs = [j-i for i,j in zip(row[:-1], row[1:])]
    return diffs
```

```python
def CheckRules(input_list):
    print("input_list", input_list)
    step_check = [1 if (0 < abs(x) < 4) else 0 for x in input_list]
    length = len(input_list)
    step_check_sum = abs(sum(step_check))
    print(step_check)
    print(step_check_sum)
    if (step_check_sum == length): #pass with flying colors
        step_ok_1 = True
        step_ok_2 = False
    elif (step_check_sum == (length-1)):
        step_ok_1 = False
        step_ok_2 = True #pass with 1 strike
    else:
        step_ok_1 = False
        step_ok_2 = False
    sign_check = [1 if (x > 0) else 0 if (x==0) else -1 for x in input_list]
    sign_check_sum = abs(sum(sign_check))
    print(sign_check)
    print(sign_check_sum)
    if (sign_check_sum == length):
        signs_ok_1 = True #pass with flying colors
        signs_ok_2 = False
    elif (sign_check_sum in [length-2, length-1]):
        signs_ok_1 = False
        signs_ok_2 = True #pass with 1 strike
    else:
        signs_ok_1 = False
        signs_ok_2 = False
    print(step_ok_1,step_ok_2)
    print(signs_ok_1,signs_ok_2)
    if step_ok_1 and signs_ok_1: #all tests good
        return True
    elif (step_ok_1 and signs_ok_2): #only one strike allowed
        return True
    elif (step_ok_2 and signs_ok_1): #only one strike allowed
        return True
    else:
        return False
```

```python
rules_ok = CheckRules([1, 2, 2, 2, 2, 2, 0])
print(rules_ok)
```

```python
df = read_df("./reports.csv")
```

```python
safe_lists = []
unsafe_list = []
unsafe_list_diffs = []
for row in df.iterrows():
    row_list = [int(x) for x in row[1].to_list() if not math.isnan(x)]
    diffs = check_row_diff(row_list)
    print("row_list", row_list)
    if CheckRules(diffs):
        print("Safe list!")
        safe_lists.append(row_list)
    else:
        print("Unsafe list")
        unsafe_list.append(row_list)
        unsafe_list_diffs.append(diffs)
```
