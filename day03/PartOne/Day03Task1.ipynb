```python
import re
```

```python
file = open("./string.txt", "r")
long_string = file.read()
```

```python
long_string
```

```python
test_sub = "mul\(\d+,\d+\)" #regexp which matches "mul(<digit>,<digit>)"
start_indices = [sub.start() for sub in re.finditer(test_sub, long_string)] #index of start of substring
end_indices = [sub.end() for sub in re.finditer(test_sub, long_string)] #index of end of substring
```

```python
for start, end in zip(start_indices, end_indices):
    numbers = long_string[(start+4):(end-1)].split(",")
    print(numbers)
```

```python
# same as above, but just as a list comprehension
substring_numbers = [long_string[(start+4):(end-1)].split(",") for start,end in zip(start_indices, end_indices)]

substring_numbers = [[int(number) for number in list] for list in substring_numbers] #convert to int

products = [numbers[0]*numbers[1] for numbers in substring_numbers]
sum(products) #answer to task one
```
