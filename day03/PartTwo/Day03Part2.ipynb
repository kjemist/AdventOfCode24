```python
import re
```

```python
file = open("./string.txt", "r")
long_string = file.read()
```

```python
long_string
```

```python
test_sub = "mul\(\d+,\d+\)" #regexp which matches "mul(<digit>,<digit>)"
start_indices = [sub.start() for sub in re.finditer(test_sub, long_string)] #index of start of substring
end_indices = [sub.end() for sub in re.finditer(test_sub, long_string)] #index of end of substring

do_indices = [sub.start() for sub in re.finditer("do\(\)", long_string)]
dont_indices = [sub.start() for sub in re.finditer("don't\(\)", long_string)]
```

```python
mul_list = []
for start, end in zip(start_indices, end_indices):
    numbers = long_string[(start+4):(end-1)].split(",")
    numbers = [int(n) for n in numbers]
    numbers.insert(0,start)
    numbers.append(mul_list)
```

```python
# same as above, but just as a list comprehension
substring_numbers = [long_string[(start+4):(end-1)].split(",") for start,end in zip(start_indices, end_indices)]

substring_numbers = [[int(number) for number in list] for list in substring_numbers] #convert to int

products = [numbers[0]*numbers[1] for numbers in substring_numbers]
sum(products) #answer to task one
```

```python
prod_indices = [[index, product] for index, product in zip(start_indices, products)] # combine the index of mul() with the associated product of its multiplication
do_indices_w_command = [[index, "do"] for index in do_indices]
dont_indices_w_command = [[index, "dont"] for index in dont_indices]
```

```python
combined_list = prod_indices + do_indices_w_command + dont_indices_w_command
combined_list.sort() #sort automatically sorts on the first index of a list
combined_list.insert(0, [1, "do"])
```

```python
combined_list
```

```python
product_sum = []
for index_product in combined_list:
    if index_product[1] == 'do':
        add =  True
    else:
        add = False
    if add:
        product_sum.add(index_product[1])
```
